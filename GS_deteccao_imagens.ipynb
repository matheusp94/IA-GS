# **Preparando saída para o modelo**

```python
import matplotlib.pyplot as plt
from inference_sdk import InferenceHTTPClient

# Inicialize o cliente
CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",
    api_key="H9ms90WhYOhBi15zG3L4"
)

# Função para realizar a inferência e mostrar resultados
def process_image(image_path, model_id="plastic-pollution-detection-av5zr/1"):
    result = CLIENT.infer(image_path, model_id=model_id)

    # Mostrar a imagem
    image = plt.imread(image_path)
    plt.imshow(image)
    plt.axis('off')  # Desativa os eixos
    plt.show()

    # Mostrar as dimensões da imagem
    width = result['image']['width']
    height = result['image']['height']
    print(f"Dimensões da imagem: {width}x{height}")

    # Mostrar as previsões
    predictions = result['predictions']
    for prediction in predictions:
        confidence = prediction['confidence']
        animal_class = prediction['class']
        print(f"Classe: {animal_class}, Confiança: {confidence}")
```
